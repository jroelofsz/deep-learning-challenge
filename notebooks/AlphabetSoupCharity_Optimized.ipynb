## Preprocessing

````python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
````

````python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])

application_df.head()
````

````python
# Determine the number of unique values in each column.
application_df.nunique()
````

````python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
````

````python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_type_count = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = application_type_count[application_type_count < 528].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
````

````python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_classification_count = application_df['CLASSIFICATION'].value_counts()

application_classification_count.head()
````

````python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = application_classification_count[application_classification_count < 1883].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
````

````python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, drop_first=True)
application_df.head()
````

````python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=59, stratify=y)
````

````python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
````

## Compile, Train and Evaluate the Model

Note, we have already compiled our initial model. This file will focus strictly on optimizing the model for the highest performance.

If you would like to view the original model, please reference the ```AlphabetSoupCharity.ipynb``` file. The model is also saved in ```/models``` as ```AlphabetSoupCharify.h5```.

Optimized Model #1 - Additional Hidden Layer, Dropout Layer, More Neurons, More Epochs.

This step includes an additional layer, more neurons, and a dropout layer that helps prevent overfitting.

````python
# declare global variables for each optimized model
input_features = len(X_train_scaled[0])
````

````python
# Define our first model to be optimized
nn_opt1 = tf.keras.models.Sequential()

# Input Layer
nn_opt1.add(tf.keras.layers.Dense(units=75, activation="relu", input_dim=input_features))
# Second Hidden Layer
nn_opt1.add(tf.keras.layers.Dense(units=50, activation="relu"))
# Third Hidden Layer
nn_opt1.add(tf.keras.layers.Dense(units=25, activation="relu"))
# Dropout Layer
nn_opt1.add(tf.keras.layers.Dropout(0.4))
# Output Layer
nn_opt1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

nn_opt1.summary()
````

````python
# Compile our new optimized model
nn_opt1.compile(loss='binary_crossentropy', optimizer="adam", metrics=["accuracy"])
````

````python
# Train our model, use shuffle to mix up the data.
fit_model = nn_opt1.fit(X_train_scaled, y_train, epochs=150, shuffle=True)
````

````python
model_loss, model_accuracy = nn_opt1.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {round(model_loss,2)}, Accuracy: {round(model_accuracy,2)}")
````

Optimized Model #2 - Keras Tuner

For this step, we will use the keras-tuner in order to tune our model's hyperparameters automatically. This will allow us to test multiple hyperparameter configurations hands free.

````python
!pip install keras-tuner
````

````python
def create_model(hp):
  # declare model
  nn_opt2 = tf.keras.models.Sequential()

  # define options for activation equation
  activation = hp.Choice("activation",["relu","tanh","sigmoid"])

  # define first layer
  nn_opt2.add(tf.keras.layers.Dense(units=hp.Int("first_units", min_value=1, max_value=10, step=2), activation=activation, input_dim = input_features))

  #define hidden layers
  for i in range(1,6):
    nn_opt2.add(tf.keras.layers.Dense(units=hp.Int("units_" + str(i), min_value=1, max_value=10, step=2), activation=activation))

  #define final layer
  nn_opt2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

  #compile model
  nn_opt2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

  return nn_opt2
````

````python
import keras_tuner as kt

# Define tuner object to search hyperparameter configurations
tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2
)
````

````python
tuner.search(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled, y_test))
````

````python
# Evalulate the top 3 models against each other.
top_3_models = tuner.get_best_models(3)
for model in top_3_models:
  model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test, verbose=2)
  print(f"Loss: {round(model_loss,2)}, Accuracy: {round(model_accuracy,2)}")
````

````python
final_model = top_3_models[0]
````
